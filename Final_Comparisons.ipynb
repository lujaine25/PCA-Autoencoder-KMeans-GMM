```python
import matplotlib.pyplot as plt
import numpy as np


# Exp 1, 3, and 5 for these numbers:
kmeans_ari = [0.6536, 0.6592, 0.7129]  
# (Exp 2, 4, 6):
gmm_ari = [0.53, 0.66, 0.39]    
# ==========================================
# 2. GENERATE PLOTS
# ==========================================
labels = ['Original Data', 'PCA Reduced', 'Autoencoder']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

# Plot K-Means vs GMM
rects1 = ax.bar(x - width/2, kmeans_ari, width, label='K-Means', color='skyblue')
rects2 = ax.bar(x + width/2, gmm_ari, width, label='GMM', color='salmon')

ax.set_ylabel('ARI Score (Higher is Better)')
ax.set_title('Final Showdown: K-Means vs. GMM Performance (ARI)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.0)
ax.legend()
ax.grid(axis='y', alpha=0.3)

# Add values on top
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```
